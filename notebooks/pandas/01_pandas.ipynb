# Python parada Análise de dados - Pandas

Será utilizada a Base de imóveis obtidas no Kaggle.

O download da base pode ser feito atravéz do [link](https://github.com/minerandodados/mdrepo/blob/master/kc_house_data.csv)

```python
# Importando o Pandas
import pandas as pd
```

```python
# Lendo uma base de dados no formato .csv.
# O parâmetro sep é usado para definir qual o separador entre os dados 
# O parametro header informa qual linha está as colunas ou se elas não existe
# Se não existe colunas (header=None) o pandas dará um número para cada atributo da base.
arquivo = 'kc_house_data.csv'
dataset = pd.read_csv(arquivo,sep=',',header=0)
```

```python
# Imprime o tipo da ariável datase
# Dataframe é uma estrutura de dados onde linhas podem ter colunas de diferentes tipos.
type(dataset)
```

```python
# Método head() imprime as 5 linhas iniciais do dataframe.
dataset.head()
```

```python
# parâmetro index_col informa a coluna na qual o datafram será indexado
dataset = pd.read_csv(arquivo,sep=',',index_col='date')
dataset.head()
```

```python
# O parâmetro usecols especifica as colunas que deve ser lidas pelo pandas, ignorando todas as outras do arquivo lido
dataset = pd.read_csv(arquivo,sep=',',usecols=['id','date','price','bedrooms'])
dataset.head()
```

```python
dataset = pd.read_csv(arquivo,sep=',',header=0)
# Imprime alem das 5 linhas
dataset.head(100)
```

```python
# O atributo columns retorna o nome das counas do dataframe
dataset.columns
```

```python
# Método count() retorna a quantidade de linhas de todas as colunas.
dataset.count()
```

```python
# Método describe() exibe informaçoes estatísticas da base de dados.
# Várias informações como desvio padrão, média valor mínimo e valor máximo de colunas
dataset.describe()
```
